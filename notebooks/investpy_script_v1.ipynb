```python
import investpy

# Fetch a list of all companies in Australia
all_companies = investpy.get_stocks_list(country='Australia')

# Define the period you want to fetch (quarterly in this case)
period = 'quarterly'

# Loop through the list of companies and fetch their cashflow data
for company in all_companies:
    stock_symbol = company
    country = 'Australia'
    
    try:
        cashflows_data = investpy.get_stock_financial_summary(stock=stock_symbol, country=country, summary_type='cash_flow_statement', period=period)
        print(f"Company: {stock_symbol}")
        print(cashflows_data)
        print("-" * 50)
    except Exception as e:
        print(f"Failed to fetch data for {stock_symbol}: {str(e)}")

```

```python
import investpy
import requests

# Define your user agent
user_agent = "Your User Agent String"

# Fetch a list of all companies in Australia
all_companies = investpy.get_stocks_list(country='Australia')

# Define the period you want to fetch (quarterly in this case)
period = 'quarterly'

# Loop through the list of companies and fetch their cashflow data
for company in all_companies:
    stock_symbol = company
    country = 'Australia'
    
    try:
        # Construct the URL for the request
        url = f"https://www.investing.com/stock-screener/Service/SearchStocks"

        # Set the headers for the request
        headers = {
            'User-Agent': user_agent,
            'Content-Type': 'application/x-www-form-urlencoded'
        }

        # Define the parameters for the request
        params = {
            'text': stock_symbol,
            'country': country
        }

        # Send the request using the requests library
        response = requests.get(url, headers=headers, params=params)

        if response.status_code == 200:
            # Parse the response data here (response.text) and extract the cashflow information.
            print(f"Company: {stock_symbol}")
            print("Cashflow data here")
            print("-" * 50)
        else:
            print(f"Failed to fetch data for {stock_symbol}. Status Code: {response.status_code}")

    except Exception as e:
        print(f"Failed to fetch data for {stock_symbol}: {str(e)}")
```
