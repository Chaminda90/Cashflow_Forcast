```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import acf, pacf
```

```python
# Load the data from CSV
df = pd.read_csv(r'C:\Users\ChamindaDilhanBISTEC\OneDrive - BISTEC Global\Cashflow forcast\data\abans_v0.1.csv') 
df['Date'] = pd.to_datetime(df['Date'])
#df.set_index('Date', inplace=True)
```

```python
# Plot the data
plt.plot(df['Cashflow_Amount'])
plt.title('Cash Flow Over Time')
plt.show()
```

```python
# Check for stationarity
result = adfuller(df['Cashflow_Amount'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
# Differencing to make the time series stationary
df_diff = df.diff().dropna()
```

```python
# ACF and PACF plots for parameter selection
acf_plot = acf(df_diff['Cashflow_Amount'], nlags=20)
pacf_plot = pacf(df_diff['Cashflow_Amount'], nlags=20)
```

```python
# Plot ACF
plt.subplot(121)
plt.stem(acf_plot)
plt.title('Autocorrelation Function')

# Plot PACF
plt.subplot(122)
plt.stem(pacf_plot)
plt.title('Partial Autocorrelation Function')

plt.show()
```

```python
import itertools

# Define the hyperparameter values to search
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 4) for x in pdq]

# Perform grid search to find the best parameters
best_aic = float("inf")
best_params = None

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = SARIMAX(df['Cashflow_Amount'], order=param, seasonal_order=param_seasonal)
            results = model.fit()
            if results.aic < best_aic:
                best_aic = results.aic
                best_params = (param, param_seasonal)
        except:
            continue

print(f"Best AIC: {best_aic}")
print(f"Best Parameters: {best_params}")
```
